```python
import numpy as np
import pandas as pd

cleaned_df = pd.read_excel('clean_data_for_ml_training.xlsx', index_col=None)
cleaned_df = cleaned_df.drop(['Unnamed: 0'], axis=1)
X = cleaned_df.drop(columns=['price_usd'])
y = cleaned_df['price_usd']
```

<h1>Bayesian optimization</h1>

```python
from skopt import BayesSearchCV
from skopt.space import Real, Integer, Categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor
import joblib
import numpy as np

def preprocess_data(X, y):
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    joblib.dump(scaler, 'pkls/scaler.pkl')
    return train_test_split(X_scaled, y, test_size=0.2, random_state=42)

def calculate_metrics(y_test, y_pred):
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    r2 = r2_score(y_test, y_pred)
    return {
        "RMSE": rmse,
        "MAE": mae,
        "MSE": mse,
        "MAPE": mape,
        "R2": r2
    }

def xgboost_bayesian_tuning(X, y):
    print("Preprocessing data...")
    X_train, X_test, y_train, y_test = preprocess_data(X, y)

    param_space = {
        'n_estimators': Integer(200, 1200),
        'max_depth': Integer(3, 15),
        'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
        'gamma': Real(0, 5),
        'subsample': Real(0.5, 1.0),
        'colsample_bytree': Real(0.5, 1.0),
        'colsample_bylevel': Real(0.5, 1.0),
        'colsample_bynode': Real(0.5, 1.0),
        'min_child_weight': Integer(1, 10),
        'reg_alpha': Real(0, 5),
        'reg_lambda': Real(0, 5),
        'max_delta_step': Integer(0, 10),
        'grow_policy': Categorical(['depthwise', 'lossguide']),
    }

    xgb_model = XGBRegressor(
        random_state=23,
        verbosity=0,
        tree_method='hist',
        n_jobs=-1
    )

    opt = BayesSearchCV(
        estimator=xgb_model,
        search_spaces=param_space,
        n_iter=30,
        scoring='neg_root_mean_squared_error',
        cv=5,
        verbose=1,
        n_jobs=-1,
        random_state=23
    )

    print("üîç Starting Bayesian optimization...")
    opt.fit(X_train, y_train)

    print("Optimization finished.")
    print("Best Parameters:", opt.best_params_)

    best_model = opt.best_estimator_
    y_pred = best_model.predict(X_test)

    metrics = calculate_metrics(y_test, y_pred)
    return metrics, best_model, opt.best_params_
```

<h1>Model implementation</h1>

```python
import matplotlib.pyplot as plt


def xgboost_tuning(X, y):
    X_train, X_test, y_train, y_test = preprocess_data(X, y)

    param_grid = {
        'n_estimators': [696],
        'max_depth': [15],
        'learning_rate': [0.052229370288901435],
        'gamma': [0],
        'subsample': [0.9545061669656212],
        'colsample_bytree': [1],
        'colsample_bylevel': [1],
        'colsample_bynode': [0.8531864051192173],
        'min_child_weight': [10],
        'reg_alpha': [1],
        'reg_lambda': [5],
        'max_delta_step': [0],
        'grow_policy': ['depthwise'],
        'verbosity': [1],
        'eval_metric': ['rmse']
    }

    param_combinations = list(product(*param_grid.values()))
    best_params = {}
    best_score = float('-inf')
    best_model = None
    best_eval_results = {}

    for params in param_combinations:
        param_dict = {key: value for key, value in zip(param_grid.keys(), params)}
        print(f"Testing parameters: {param_dict}")

        model = XGBRegressor(**param_dict, random_state=42, early_stopping_rounds=10)
        model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], verbose=False)

        score = model.score(X_test, y_test)

        if score > best_score:
            best_score = score
            best_model = model
            best_params = param_dict
            best_eval_results = model.evals_result()

    # Plot training and test loss curves
    plt.figure(figsize=(8, 5))
    plt.plot(best_eval_results['validation_0']['rmse'], label='Train Loss (RMSE)')
    plt.plot(best_eval_results['validation_1']['rmse'], label='Test Loss (RMSE)')
    plt.xlabel("Iteration")
    plt.ylabel("Loss (RMSE)")
    plt.title("Training and Test Loss Over Iterations")
    plt.legend()
    plt.show()

    y_pred = best_model.predict(X_test)
    return calculate_metrics(y_test, y_pred), best_model, best_params
```

```python
xgboost_metrics, xgboost_best_model, xgboost_best_params = xgboost_tuning(X, y)
```

```python
import pandas as pd
import joblib
import xgboost as xgb
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

def prepare_input(incoming_data: pd.DataFrame, expected_columns: list) -> pd.DataFrame:
    """One-hot encodes and aligns input data with the trained model's feature set."""
    categorical_cols = ['district_name', 'foundation_name', 'layout_name', 'repair_name', 'wc_name']
    incoming_encoded = pd.get_dummies(incoming_data, columns=categorical_cols)

    # Ensure all expected columns exist
    for col in expected_columns:
        if col not in incoming_encoded.columns:
            incoming_encoded[col] = 0

    # Drop any extra columns and ensure correct order
    incoming_encoded = incoming_encoded[expected_columns].astype(float)  

    return incoming_encoded

def predict_with_confidence(model: xgb.XGBRegressor, X_row: pd.DataFrame) -> dict:
    """Generates prediction with confidence interval using XGBoost trees."""
    booster = model.get_booster()  
    dmatrix = xgb.DMatrix(X_row)

    # Use num_boosted_rounds() instead of best_ntree_limit
    num_trees = booster.num_boosted_rounds()  

    # Extract predictions from individual trees using iteration_range
    all_preds = np.array([booster.predict(dmatrix, iteration_range=(i, i+1))[0] for i in range(num_trees)])

    mean_pred = np.mean(all_preds)
    std_pred = np.std(all_preds)
    lower_bound = mean_pred - 1.96 * std_pred  # 95% CI approximation
    upper_bound = mean_pred + 1.96 * std_pred  

    return {
        'prediction': mean_pred,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'std_dev': std_pred,
        'all_pred': all_preds.tolist()
    }

# Load trained model and preprocessing tools
model = xgboost_best_model
scaler = joblib.load('pkls/scaler.pkl')
expected_columns = X.columns

# Incoming data
incoming_data = pd.DataFrame([{
    "district_name": "–ß–∏–ª–∞–Ω–∑–∞—Ä—Å–∫–∏–π —Ä–∞–π–æ–Ω",
    "number_of_rooms": "2",
    "floors": "4",
    "total_floors": "4",
    "total_area": "20",
    "foundation_name": "–ü–∞–Ω–µ–ª—å–Ω—ã–π",
    "layout_name": "–°–º–µ–∂–Ω–æ-—Ä–∞–∑–¥–µ–ª—å–Ω–∞—è",
    "wc_name": "2 —Å–∞–Ω—É–∑–ª–∞ –∏ –±–æ–ª–µ–µ",
    "repair_name": "–ï–≤—Ä–æ—Ä–µ–º–æ–Ω—Ç",
    "year": 2024,
    "month":11,
    "is_primary": "1"
}])


# Preprocess input
processed_input = prepare_input(incoming_data, expected_columns)
scaled_input = scaler.transform(processed_input)  

# Make prediction with confidence interval
result = predict_with_confidence(model, scaled_input)

# Output results
print(f"Prediction: {result['prediction']:.2f}")
print(f"Confidence Interval (approx. 95%): {result['lower_bound']:.2f} to {result['upper_bound']:.2f}")
print(f"Standard Deviation: {result['std_dev']:.2f}")

# Visualizing the distribution of tree-based predictions
plt.hist(result['all_pred'], bins=15, edgecolor='black')
plt.xlabel('Predicted Value')
plt.ylabel('Frequency')
plt.title('Distribution of Model Predictions Across Trees')
plt.show()
```
