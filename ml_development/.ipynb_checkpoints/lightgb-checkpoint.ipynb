```python
import numpy as np
import pandas as pd

cleaned_df = pd.read_excel('clean_data_for_ml_training.xlsx', index_col=None)
cleaned_df = cleaned_df.drop(['Unnamed: 0'], axis=1)
X = cleaned_df.drop(columns=['price_usd'])
y = cleaned_df['price_usd']
```

<h1>Bayesian optimization</h1>

```python
from lightgbm import LGBMRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from skopt import BayesSearchCV
from skopt.space import Real, Integer
import numpy as np
import matplotlib.pyplot as plt

def preprocess_data(X, y):
    print("[INFO] Starting data preprocessing...")
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    print("[INFO] Data preprocessing completed.")
    return X_scaled, y

def calculate_metrics(y_test, y_pred):
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    r2 = r2_score(y_test, y_pred)
    return {
        "RMSE": rmse,
        "MAE": mae,
        "MSE": mse,
        "MAPE": mape,
        "R2": r2
    }

def bayesian_lightgbm_tuning_with_loss_plot(X, y):
    print("[INFO] Starting Bayesian optimization with cross-validation...")
    X_scaled, y_scaled = preprocess_data(X, y)

    param_space = {
        'n_estimators': Integer(100, 1000),
        'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
        'max_depth': Integer(-1, 15),  # -1 will represent 'None' for unlimited depth
        'num_leaves': Integer(20, 500),
        'min_child_samples': Integer(5, 50),
        'min_child_weight': Real(1e-5, 1e-1, prior='log-uniform'),
        'subsample': Real(0.5, 1.0),
        'colsample_bytree': Real(0.5, 1.0),
        'reg_alpha': Real(0.0, 1.0),
        'reg_lambda': Real(0.0, 1.0),
    }

    model = LGBMRegressor(random_state=23, force_col_wise=True)

    opt = BayesSearchCV(
        estimator=model,
        search_spaces=param_space,
        n_iter=30,
        scoring='neg_root_mean_squared_error',
        cv=5,
        verbose=0,
        n_jobs=-1,
        random_state=23
    )

    print("[INFO] Running Bayesian hyperparameter search...")
    opt.fit(X_scaled, y_scaled)
    print("[INFO] Bayesian optimization completed.")

    best_params = opt.best_params_
    if best_params.get('max_depth') == -1:
        best_params['max_depth'] = None  # Convert -1 to None for LightGBM

    print(f"[INFO] Best hyperparameters found: {best_params}")

    print("[INFO] Splitting data into train and test sets for final training...")
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

    print("[INFO] Training final model with best hyperparameters and early stopping...")
    best_model = LGBMRegressor(
        random_state=23,
        force_col_wise=True,
        **best_params,
        early_stopping_rounds=10
    )

    best_model.fit(
        X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        eval_metric='rmse'
    )

    print("[INFO] Training completed. Extracting training history...")

    eval_results = best_model.evals_result_
    train_loss = eval_results['training']['rmse']
    val_loss = eval_results['valid_1']['rmse']

    print("[INFO] Plotting training and validation RMSE over iterations...")
    plt.figure(figsize=(10, 6))
    plt.plot(train_loss, label='Train RMSE')
    plt.plot(val_loss, label='Validation RMSE')
    plt.xlabel('Iterations')
    plt.ylabel('RMSE')
    plt.title('Training and Validation RMSE over iterations')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print("[INFO] Predicting on test data and calculating metrics...")
    y_pred = best_model.predict(X_test)
    metrics = calculate_metrics(y_test, y_pred)

    print(f"[INFO] Metrics on test set: {metrics}")

    return metrics, best_model, best_params
```

<h1>Model implementation</h1>

```python
from itertools import product
import numpy as np
from lightgbm import LGBMRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import joblib
# Function to preprocess data
def preprocess_data(X, y):
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    joblib.dump(scaler, 'pkls/scaler.pkl')
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    return X_train, X_test, y_train, y_test

# Function to calculate metrics
def calculate_metrics(y_test, y_pred):
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    r2 = r2_score(y_test, y_pred)
    return {
        "RMSE": rmse,
        "MAE": mae,
        "MSE": mse,
        "MAPE": mape,
        "R2": r2
    }

param_grid = {
    'n_estimators': [996], 
    'learning_rate': [0.11213619582835252], 
    'max_depth': [0], 
    "num_leaves": [273],
    'min_child_weight': [0.012133013204476073],
    "min_child_samples": [33],
    'subsample': [0.7248500765862023], 
    'colsample_bytree': [0.9517624347879201], 
    'reg_alpha': [1], 
    'reg_lambda': [0.9630469936739557], 
    'early_stopping_rounds': [10], 
    'random_state': [23] 
}



def lightgbm_tuning(X, y):
    X_train, X_test, y_train, y_test = preprocess_data(X, y)

    param_combinations = list(product(
        param_grid['n_estimators'], param_grid['learning_rate'], param_grid['max_depth'], 
        param_grid['num_leaves'], param_grid['min_child_samples'], param_grid['min_child_weight'], 
        param_grid['subsample'], param_grid['colsample_bytree'], param_grid['reg_alpha'], 
        param_grid['reg_lambda'], param_grid['early_stopping_rounds']
    ))

    best_params = {}
    best_score = float('-inf')
    best_model = None
    best_eval_results = {}

    for params in param_combinations:
        (n_estimators, learning_rate, max_depth, num_leaves, min_child_samples, min_child_weight, 
         subsample, colsample_bytree, reg_alpha, reg_lambda, early_stopping_rounds_value) = params

        model = LGBMRegressor(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            max_depth=max_depth,
            num_leaves=num_leaves,
            min_child_samples=min_child_samples,
            min_child_weight=min_child_weight,
            subsample=subsample,
            colsample_bytree=colsample_bytree,
            reg_alpha=reg_alpha,
            reg_lambda=reg_lambda,
            early_stopping_rounds=early_stopping_rounds_value,
            random_state=23,
            force_col_wise=True,
        )
        
        model.fit(
            X_train, y_train,
            eval_set=[(X_train, y_train), (X_test, y_test)],
            eval_metric="rmse"
        )

        score = model.score(X_test, y_test)  
        
        if score > best_score:
            best_score = score
            best_model = model
            best_params = params
            best_eval_results = model.evals_result_  

    plt.figure(figsize=(8, 5))
    plt.plot(best_eval_results['training']['l2'], label='Train Loss')
    plt.plot(best_eval_results['valid_1']['l2'], label='Test Loss')
    plt.xlabel("Iteration")
    plt.ylabel("Loss (RMSE)")
    plt.title("Training and Test Loss Over Iterations")
    plt.legend()
    plt.show()

    y_pred = best_model.predict(X_test)
    return calculate_metrics(y_test, y_pred), best_model, best_params, X_train

```

```python
lightgbm_metrics, lightgbm_best_model, lightgbm_best_params, X_train = lightgbm_tuning(X, y)
```

```python
import pandas as pd
import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

def preprocess_data_single_entry(X, scaler):
    X_scaled = scaler.transform(X)
    return X_scaled

def prepare_input(incoming_data: pd.DataFrame, expected_columns: list) -> pd.DataFrame:
    categorical_cols = ['district_name', 'foundation_name', 'layout_name', 'repair_name', 'wc_name']
    incoming_encoded = pd.get_dummies(incoming_data, columns=categorical_cols)

    # Add missing columns (ensuring alignment with expected_columns)
    for col in expected_columns:
        if col not in incoming_encoded.columns:
            incoming_encoded[col] = 0

    # Retain only the expected columns
    incoming_encoded = incoming_encoded[expected_columns]

    # Ensure correct column order
    incoming_encoded = incoming_encoded.astype(int)

    return incoming_encoded

def predict_with_confidence(model, X_row: pd.DataFrame) -> dict:
    booster = model.booster_
    all_preds = [booster.predict(X_row, num_iteration=tree_idx)[0] for tree_idx in range(booster.current_iteration())]
    
    mean_pred = np.mean(all_preds)
    std_pred = np.std(all_preds)
    
    return {
        'prediction': mean_pred,
        'lower_bound': mean_pred - 1.96 * std_pred,
        'upper_bound': mean_pred + 1.96 * std_pred,
        'std_dev': std_pred,
        'all_pred': all_preds
    }

# Load the trained LightGBM model and scaler
model = lightgb_best_model
scaler = joblib.load("pkls/scaler.pkl")

expected_columns = X.columns

# Incoming apartment data
incoming_data = pd.DataFrame([{
    "district_name": "Чиланзарский район",
    "number_of_rooms": "5",
    "floors": "1",
    "total_floors": "5",
    "total_area": "120",
    "foundation_name": "Панельный",
    "layout_name": "Смежно-раздельная",
    "wc_name": "2 санузла и более",
    "repair_name": "Евроремонт",
    "year": 2024,
    "month":11,
    "is_primary": "1"
}])

processed_input = prepare_input(incoming_data, expected_columns)
scaled_input = preprocess_data_single_entry(processed_input, scaler)

result = predict_with_confidence(model, scaled_input)

print(f"Prediction: {result['prediction']:.2f}")
print(f"Confidence Interval (approx. 95%): {result['lower_bound']:.2f} to {result['upper_bound']:.2f}")
print(f"Standard Deviation: {result['std_dev']:.2f}")

plt.hist(result['all_pred'], bins=10, edgecolor='black')
plt.xlabel('Predicted Value')
plt.ylabel('Frequency')
plt.title('Histogram of Model Predictions')
plt.show()
```
